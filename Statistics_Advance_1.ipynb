Question 1: What is a random variable in probability theory?

 - A random variable is a numerical quantity whose value depends on the outcome of a random event. It maps the outcomes of a random experiment to numerical values.

   Example: When tossing a coin, let X = 1 if heads appear and X = 0 if tails appear.
   So, X is a random variable.

Question 2: What are the types of random variables?

 - There are two main types of random variables:

   Discrete Random Variable ‚Äì Takes on countable values (e.g., 0, 1, 2, 3).
   Example: Number of students present in a class.

   Continuous Random Variable ‚Äì Takes on infinitely many possible values within an interval.
   Example: Height, weight, or time taken to finish a race.

Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title('Normal Distribution (Mean=50, SD=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.

 ‚óè Write the Python code to compute the mean sales and its confidence interval.


 - Step 1: Compute sample mean and standard deviation
   The CLT allows us to assume the sample mean follows a normal distribution.
   We can use the formula:

   CI=ùëãÀâ¬±ùëç0.975√óùë†ùëõCI=XÀâ¬±Z0.975√óns

```python
import numpy as np
from scipy import stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Compute mean and standard deviation
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

# Compute 95% confidence interval
confidence = 0.95
z = stats.norm.ppf(1 - (1 - confidence)/2)
margin_of_error = z * (std_sales / np.sqrt(n))
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
